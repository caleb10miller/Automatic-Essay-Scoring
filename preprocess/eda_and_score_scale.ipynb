# Explanation for Grading Scales

## Set 1 

ranges from 2 - 12\
Each score is a whole number - no need to account for grades like 10.5

Max value is 12 ---> 11 <= score <= 12 represents A\
75th percentile is 10 ---> 9 <= score <= 10 represents B\
The average value (mean and median) is approximately 8 ---> score of 8 = C\
25th percentile is 8 (same as average median) ---> 6 <= score <= 7 represents D\
Minimum value is 2 (outlier) ---> 2 <= score <= 5 represents F

## Set 2

ranges from 1 - 6\
Each score is a whole number - no need to account for grades like 3.5

Max value is 6 ---> 5 <= score <= 6 represents A\
75th percentile is 4 ---> 4 represents B\
Average value (mean and median) is approximately 3 ---> 3 represents C\
25th percentile is 3 (same as median) ---> 2 represents D\
Min value is 1 ---> 1 represents F

## Set 3

ranges from 0 - 3\
Each score is a whole number - no need to account for grades like 1.5

Max value is 3 ---> 3 represents A\
75th percentile is 2 ---> 2 represents B\
Center of data is between 1 and 2 (impossible) ---> No grade represents C (only 4 grades possible)\
25th percentile is 1 ---> 1 represents D\
Min value is 0 ---> 0 represents F

```python
# Dealing with NaN in scoring column

df.domain1_score.isna().sum()
```

1 nan present - need to remove it

```python
df.dropna(subset=['domain1_score'], inplace=True)
df.domain1_score.isna().sum()
```

NaN values dealt with

```python
def scale_essay_set1(df):
    score_mapping = {
        'A': (11,12),
        'B': (9,10),
        'C': (8,8),
        'D': (6,7),
        'F': (2,5)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
def scale_essay_set2(df):
    score_mapping = {
        'A': (5,6),
        'B': (4,4),
        'C': (3,3),
        'D': (2,2),
        'F': (1,1)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
def scale_essay_set3and4(df):
    score_mapping = {
        'A': (3,3),
        'B': (2,2),
        'C': (1,1),
        'D': (0,0)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
def scale_essay_set5and6(df):
    score_mapping = {
        'A': (4,4),
        'B': (3,3),
        'C': (2,2),
        'D': (1,1),
        'F': (0,0)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
def scale_essay_set7(df):
    score_mapping = {
        'A': (21,24),
        'B': (18,20),
        'C': (14,17),
        'D': (9,13),
        'F': (2,8)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
def scale_essay_set8(df):
    score_mapping = {
        'A': (49,60),
        'B': (40,48),
        'C': (36,39),
        'D': (26,35),
        'F': (10,25)
    }
    def map_score_to_grade(score):
        for grade, (lower_bound, upper_bound) in score_mapping.items():
            if lower_bound <= score <= upper_bound:
                return grade
        return None
    
    df['scaled_grade'] = df['domain1_score'].apply(map_score_to_grade)

    return df
```

```python
scaled_df = pd.DataFrame()

for s in df.essay_set.unique():
    if (s == 1):
        scaled = scale_essay_set1(df[(df.essay_set == s)])
    elif (s == 2):
        scaled = scale_essay_set2(df[(df.essay_set == s)])
    elif ((s == 3) | (s == 4)):
        scaled = scale_essay_set3and4(df[(df.essay_set == s)])
    elif ((s == 5) | (s == 6)):
        scaled = scale_essay_set5and6(df[(df.essay_set == s)])
    elif (s == 7):
        scaled = scale_essay_set7(df[(df.essay_set == s)])
    elif (s == 8):
        scaled = scale_essay_set8(df[(df.essay_set == s)])
    
    scaled_df = pd.concat([scaled_df, scaled], ignore_index=True)
scaled_df
```
